```python
# https://stackoverflow.com/questions/38852822/how-to-flatten-xml-file-in-python
# Original code: https://codereview.stackexchange.com/a/21035
from collections import OrderedDict

def flatten_dict(d):
    def items():
        for key, value in d.items():
            if isinstance(value, dict):
                for subkey, subvalue in flatten_dict(value).items():
                    yield key + "." + subkey, subvalue
            else:
                yield key, value

    return OrderedDict(items())

import xmltodict

# Convert to dict
with open('../data/regzam_short.xml', 'rb') as f:
    xml_content = xmltodict.parse(f)

# Flatten dict
flattened_xml = flatten_dict(xml_content)

# Print in desired format
for k,v in flattened_xml.items():
    print('{} = {}'.format(k,v))


```

How to flatten xml into dict:

```python
from collections import defaultdict

def xml_to_dict(root):
    # Create a defaultdict to store the result
    result = defaultdict(list)

    # Iterate over the root element's children
    for child in root:
        # Get the child's tag name and text
        tag = child.tag
        text = child.text

        # If the child has children, recursively process them
        if child.getchildren():
            sub_dict = xml_to_dict(child)
            for k, v in sub_dict.items():
                result[tag + '_' + k].extend(v)
        else:
            # If the child has no children, store its tag and text in the result
            result[tag].append(text)

    return result

# Example XML document
xml = """
<root>
    <a>1</a>
    <b>2</b>
    <c>
        <d>
          <f>3</f>
        </d>
        <e>4</e>
    </c>
</root>
"""


# Parse the XML document into an ElementTree object
root = etree.fromstring(xml)

# Convert the ElementTree object to a dictionary
result = xml_to_dict(root)

# Print the result
print(result)
```
